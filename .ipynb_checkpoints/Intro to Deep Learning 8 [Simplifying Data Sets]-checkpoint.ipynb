```python
#Build PCA - Principal Component Analysis
#Why do we need Dimensionality Reduction(DR) ?
###Reason 1 - space efficiency
###Reason 2 - computational efficiency
###Reason 3 - visualisation purpose (high dimensional visualisation is not possible by human eye)

#Build PCA and compare it to T-SNE and LDA (3 DRs in total)
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(1) #for debugging purpose since the fixing of seed f
#ixes the starting point of random number generation
```

```python
#Step 1 - Create our dataset

mu_vec1 = np.array([0,0,0]) #sample mean
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]]) #sample covariance

class1_sample = np.random.multivariate_normal(mu_vec1,cov_mat1,20).T
print class1_sample
#3*20 matrix (3 columns and 20 rows)
```

```python
mu_vec2 = np.array([1,1,1]) #sample mean
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]]) #sample covariance

class2_sample = np.random.multivariate_normal(mu_vec2,cov_mat2,20).T
print class2_sample
#3*20 matrix (3 columns and 20 rows)
```

```python
#plot samples
%pylab inline

#plotting
#figure, width and height
fig = plt.figure(figsize=(15,15))
#3d subplot
#These are subplot grid parameters encoded as a single integer, so 1x1 grid 1st subplot
ax = fig.add_subplot(111, projection='3d')
#fontsize
plt.rcParams['legend.fontsize'] = 10

#plot samples
ax.plot(class1_sample[0,:], class1_sample[1,:], class1_sample[2,:],
        'o', markersize=8, color='blue', alpha=0.5, label='class1')
ax.plot(class2_sample[0,:], class2_sample[1,:], class2_sample[2,:],
        '^', markersize=8, alpha=0.5, color='green', label='class2')

ax.legend(loc='upper right')

plt.show()
```

```python
#step 1 : take whole dataset ignoring classes
#make it one big dataset 3*40 (since each were 3*20 and 3*30)
all_samples = np.concatenate((class1_sample,class2_sample),axis=1)
```

```python
all_samples
```

```python
all_samples.T
```

```python
# step 2. compute the d dimensional mean vector, to help compute covariance matrix
```

```python
#mean for each feature
mean_x = np.mean(all_samples[0,:])
mean_y = np.mean(all_samples[1,:])
mean_z = np.mean(all_samples[2,:])

#3D mean vector
mean_vector = np.array([mean_x,mean_y,mean_z])
mean_vector
```

```python
df = pd.DataFrame(all_samples.T,columns=['x','y','z'])
```

```python
mean(df).values
```

```python
# step 3. compute the covariance matrix
#covariacne matrix models relationship between variables. 
#Variance is the degree by which a random vairable changes with respect to its expected value
#Covariance is the degree by which two different random variables change with respect to each other. 
#measures relationship between each feature
```

```python
df.cov().values
```

```python
cov_matrix = np.cov([all_samples[0,:],all_samples[1,:],all_samples[2,:]])
```

```python
cov_matrix
```

```python
# Step 4. computer eigenvectors and eigenvalues calculation

#Eigenvalues/vectors are instrumental to understanding electrical circuits, mechanical systems, ecology and 
#even Google's PageRank algorithm. 
#Eigenvectors make understanding linear transformations easy.
#They are the "axes" (directions) along which a linear transformation acts simply by 
#"stretching/compressing" and/or "flipping"; 
#eigenvalues give you the factors by which this compression occurs.
#There are a lot of problems that can be modeled with linear transformations, and 
#the eigenvectors give very simply solution
#The more directions you have along which you understand the behavior of a 
#linear transformation, the easier it is to understand the linear transformation; 
#so you want to have as many linearly independent eigenvectors as possible associated 
#to a single linear transformation.


#process : the eigenvector associated with the largest eigen value gives the
#axes/direction along the dataset having the highest variation. 2nd highest towards the 2nd highest variation
#an so on. This was while choosing principal components we choose these principal vectors i.e. principal axes 
#as our new axes in decreasing order of explaining the variance of our data. Once the number of axes from these 
#principal vectors are chosen, then the data is linearly transformed from current axes to the new axis by 
#perpendicular dropping on the new axes.
```

```python
eigen_values,eigen_vectors = np.linalg.eig(cov_matrix)
for i,j in zip(eigen_values,eigen_vectors.T):
    print "Eigen Value {0} for Eigen Vector {1}".format(i,j)
```

```python
# step 5. sort eigenvector by decreasing value
```

```python
eig_vect_sorted_dec_order = eigen_vectors[:,np.argsort(eigen_values)[::-1]]
```

```python
eig_vect_sorted_dec_order
```

```python
eigen_values[np.argsort(eigen_values)[::-1]]
```

```python
# now choose the number of new dimensions < earlier number of dimensions(here 3)
# say 2
#so we choose 2 top eigen vectors and create the matrix which when multiplied with the initial data then the
#dimensionality of the data reduces from 3 to 2
#How ? Linear Algebra / Matrix Multiplication
new_axes = eig_vect_sorted_dec_order[:,:2]
```

```python
new_axes
```

```python
new_axes.shape
```

```python
# step 6. use d x k to transform samples to new subspace
```

```python
transformed = np.dot(new_axes.T,all_samples)
```

```python
plt.plot(transformed[0,0:20], transformed[1,0:20],
         'o', markersize=7, color='green', alpha=0.5, label='class1')
plt.plot(transformed[0,20:40], transformed[1,20:40],
         '^', markersize=7, color='red', alpha=0.5, label='class2')
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.title('Transformed samples with class labels')

plt.show()
```
